In this project, my goal was to be able to predict the performances of Triple Jumpers at the Atlantic 10 Indoor Conference. Using TFFRS.org I made a spreadsheet of previous years performances at the A10 Indoor Conference Meet. I used the 2017, 2018, 2019, 2020, and 2022 meets. I gathered the name and school, Place, Mark at Conference, and the athlete's previous 4 jumps from that indoor season. If there was any "No Mark's" or an athlete had not performed 4 times previous to the conference meet, a 0 was inserted for their jump.

I first imported pandas and all of the TSV files. I randomly selected 3 of the datasets to be training data, 1 dataset to be development data, and 1 dataset to be test data. 2022, 2020, and 2017 were selected to be training data, 2019 was selected to be the development set, and 2018 was selected to be the test set. All of my sources that I referred to are commented in the code where they were used.

```python
#Day 7 Lab
import pandas as pd
j2017 = pd.read_table('2017.tsv')
test = pd.read_table('2018.tsv')
dev = pd.read_table('2019.tsv')
j2020 = pd.read_table('2020.tsv')
j2022 = pd.read_table('2022.tsv')

#https://jakevdp.github.io/PythonDataScienceHandbook/03.06-concat-and-append.html
training = pd.concat([j2017, j2020, j2022])
training
```

Here I concatanated the training data into one dataframe.

```python

ytrain=training.ConfJump
xtrain = training.drop(['Name','Place','ConfJump'], axis=1)
xtrain


```

I then seperated my training data into two different dataframes. One is the predictors for the decision tree regressor and the other is the target for the decision tree regressor. The predictor dataframe is the 4 jumps previous to the conference meet and the target is the Conference Meet Jumps.

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html

from sklearn.tree import DecisionTreeRegressor
final_model=DecisionTreeRegressor(max_leaf_nodes=30,random_state=0)
final_model.fit(xtrain,ytrain)
final_model.predict(xtrain)
fp=final_model.predict(xtrain)
```

I then imported the decision tree regressor from scikit learn. I fit the training data to the decision tree regressor.

I set the # of nodes on the decision tree regressor to what made the development set the best.

```python
training.join(pd.DataFrame(fp, columns = ['Predicted']))
```

I joined the output of the Decision Tree Regressor as a column called "Predicted".

```python

xdev = dev.drop(['Name','Place','ConfJump'], axis=1)


final_model.predict(xdev)
devresults=final_model.predict(xdev)
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html
dev.join(pd.DataFrame(devresults, columns = ['Predicted']))
```

I then ran the model on the development set. I adjusted the number of nodes and reran the dataset to find the optimal number of nodes.


```python
xtest = test.drop(['Name','Place','ConfJump'], axis=1)


final_model.predict(xtest)
testresults=final_model.predict(xtest)
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html
testfinal =test.join(pd.DataFrame(testresults, columns = ['Predicted']))
testfinal
```

After finding the optimal number of nodes using the development set, I predicted the test set with the model.

```python
pabsolutedifference = abs(testfinal['ConfJump'] - testfinal['Predicted'])
testfinal.join(pd.DataFrame(absolutedifference, columns = ['Absolute Difference']))
```

The Regressor was able to predict 4/8 of the places correctly. The absolute difference column that I added helps show how much people may have under or overperformed compared to the rest of their indoor season. In Nathan Adeyemi's case, the regressor predicted him at a much higher place but he performed much worse compared to his other jumps that season. Because of this, his absolute difference is very high. In conclusion this model was able to somewhat accurately predict the outcome of the A10 Indoor Championship Triple Jump Final. However, it is not perfect as it could only guess 50% of the places correctly.
